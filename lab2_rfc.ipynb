```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from time import time
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("abalone.csv")
print("Number of samples: %d" % len(data))
data.head()
```

```python
data["M"] = np.nan
data["F"] = np.nan
data["I"] = np.nan
```

```python
for i in range (len(data["Sex"])):
    data["M"] = data["Sex"] == "M"
    data["F"] = data["Sex"] == "F"
    data["I"] = data["Sex"] == "I"
del data["Sex"]
```

```python
data[["M", "F", "I"]] *= 1
```

```python
x = data.drop(columns=(["Rings"]))
y = data["Rings"]
```

```python
X = x.to_numpy(dtype='float')
Y = y.to_numpy(dtype='float')
# x = data.values.astype(np.float)
# y = data.values.astype(np.float)
from sklearn.utils import shuffle
#перетасовка выборки
X, Y = shuffle(X, Y, random_state=0)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import cross_val_score

scores = []
x_scores = []

def arith_mean(lst):
   return float(sum(lst)) / len(lst)

tin = time()
for i in range(1, 50):
    rfc = RandomForestRegressor(max_depth = i)
    score = cross_val_score(rfc, X, Y, scoring='r2', cv = 5)
    scores.append(arith_mean(score))
    x_scores.append(i)
tout = time()

display(str(round((-tin+tout)/(50),3)) + " сек")
```

```python
%matplotlib notebook
plt.ylabel("R2")
plt.xlabel("Depth")
display(plt.plot(x_scores,scores))
```

```python
scores2 = []
x_scores = []

def arith_mean(lst):
   return float(sum(lst)) / len(lst)

tin = time()
for i in range(1, 50):
    rfc = RandomForestRegressor(n_estimators = i, max_depth = 11)
    score = cross_val_score(rfc, X, Y, scoring='r2', cv = 5)
    scores2.append(arith_mean(score))
    x_scores.append(i)
tout = time()
display(str(round((-tin+tout)/(50),3)) + " сек")
```

```python
%matplotlib notebook
plt.ylabel("R2")
plt.xlabel("Depth")
display(plt.plot(x_scores,scores2))
```

```python
from sklearn.ensemble import GradientBoostingRegressor

score_gbr = []
score_gbr2 = []
accuracy_gb = [1, 0.5, 0.3, 0.2, 0.1]
for i in accuracy_gb:
    gbr = GradientBoostingRegressor(learning_rate = i, n_estimators=100, random_state = 17)
    gbr.fit(x_train, y_train)
    score_gbr.append(gbr.score(x_test, y_test))
display(score_gbr)
```

```python
%matplotlib notebook
plt.xlabel("learning_rate")
plt.ylabel("R2")
plt.plot(accuracy_gb,score_gbr)
```

```python
# staged_predict

from sklearn.metrics import r2_score

score_r2_test = []
score_r2_train = []

gbr = GradientBoostingRegressor(learning_rate = 0.1, n_estimators = 100)
gbr.fit(x_train, y_train)

for i, y_pred in enumerate(gbr.staged_predict(x_test)):
    score_r2_test.append(r2_score(y_test, y_pred))
for i, y_pred1 in enumerate(gbr.staged_predict(x_train)):
    score_r2_train.append(r2_score(y_train, y_pred1))
```

```python
%matplotlib notebook
plt.plot(score_r2_test)
plt.plot(score_r2_train)
plt.xlabel("staged_predict")
plt.ylabel("R2")
```
